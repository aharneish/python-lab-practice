1. Use `matplotlib.pyplot.plot` to produce a plot of the functions 
   $f(x) = e^{-x/10}sin(πx)$ and $g(x) = xe^{−x/3}$ over the interval [0, 10]. 
   Include labels for the x- and y-axes, and a legend explaining which line is which plot.
   Save the plot as a .jpg (“Jpeg”).


```python
import matplotlib.pyplot as plt
import numpy as np
f=lambda x: np.exp(-x/10)*np.sin(np.pi*x)
g=lambda x: x*np.exp(-x/3)
x=np.linspace(0,10,100)
F=f(x)
G=g(x)

plt.figure(figsize=(10,6))
plt.ylabel("x",fontsize=18)
plt.xlabel("F(x)",fontsize=18)
plt.plot(x,F,label="f(x)=e^(-x/10)*sin(pi*X)")
plt.legend()
plt.show()

plt.figure(figsize=(10,6))
plt.xlabel("x",fontsize=18)
plt.ylabel("G(x)",fontsize=18)
plt.plot(x,G,label="g(x)=x*e^(-x/3)")
plt.legend()
plt.show()
```

2. Compare plots of $y = x^3$ for $-10 \le x \le 10$ using a logarithmic scale on the x-axis, y-axis and both axes.
   What is the difference between using `ax.set_xscale('log')` and `ax.set_xscale('symlog')`?

```python
from math import * 
import matplotlib.pyplot as plt
import numpy as np
f=lambda x: x**3
x=np.linspace(-10,10,100)
F=f(x)
plt.figure(figsize=(10,10))
plt.subplot(2,3,5)
plt.semilogx(x,F) # semilogx plot function
plt.title('Semilogx')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
F=f(x)
plt.figure(figsize=(10,10))
plt.semilogy(x,F) # semilogx plot function
plt.title('Semilogy')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
F=f(x)
plt.figure(figsize=(10,10))
plt.subplot(3,3,5)
plt.loglog(x,F) # semilogx plot function
plt.title('loglog')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
```

4. Use `matplotlib.pyplot.plot` to produce the surface and contour plot of the function $f(X) = x_1^2x_2^2(1-x_1-x_2).$ Can you identify maxima and minima using cantour plots. For better visualization, make interactive suface plot.

```python
import matplotlib.pyplot as plt
import numpy as np
X=lambda a,b: (a**2)*(b**2)*(1-a-b)
x=np.linspace(0,10,100)
y=x.copy()
f=X(x,y)
plt.plot(f,label="$x1^2*x2^2(1-x1-x2)$")
plt.title('f(x)=$x1^2*x2^2(1-x1-x2)$')
plt.xlabel('$x1$')
plt.ylabel('$x2$')
plt.grid()
plt.show()
fig= plt.figure(figsize = (12,10))
ax = plt.axes(projection='3d')

x = np.linspace(-100,100,100)
y = x.copy()
x1, x2 = np.meshgrid(x,y)
Z = (x1**2)*(x2**2)*(1-x1-x2)
plt.title('Plot: $f(x) = x1^2*x2^2*(1-x1-x2)$', fontsize=18)
surf = ax.plot_surface(x1,x2,Z, cmap= plt.cm.inferno)
# set axes label
ax.set_xlabel('x', labelpad=20, fontsize=18)
ax.set_ylabel('y', labelpad=20, fontsize=18)
ax.set_zlabel('z', labelpad=7, fontsize=18)

fig.colorbar(surf, shrink=0.5, aspect=8) # adding colorbar

plt.savefig('plot5.jpeg',dpi=300)
plt.show()
```

5. Plot gamma dribution for the parameters $(\alpha,\lambda),\ \alpha = 1, 2, 9; \ \theta = 0.5,1$

```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
alpha=[1,2,9]
lamda=[0.5,1]
x=symbols('x')
gamma=[]
for i in range(0,len(alpha)):
    i=x**(alpha[i])*exp(-x)
    expr=integrate(i,(x,0,oo))
    gamma.append(expr)
g1=[]
for i in range(0,len(alpha)):
    init1 = ((lamda[0]**alpha[i])/gamma[i])*(x**(alpha[i]-1))*(exp(-lamda[0]*x))
    g1.append(integrate(init1 , (x,0,oo)))
g2 = []
# for lambda value = 1
for i in range(0,len(alpha)):
    init2 = ((lamda[1]**alpha[i])/gamma[i])*(x**(alpha[i]-1))*(exp(-lamda[1]*x))
    g2.append(integrate(init2 , (x,0,oo)))

plt.figure(figsize=(16,5))  
plt.suptitle('Gamma Function with parameters alpha and lambda',fontsize=20)
plt.subplot(1,2,1)    
plt.scatter(alpha,g1, c='r', label = 'lambda=0.5')
plt.plot(alpha,g1, c='b')
plt.xlabel('alpha',fontsize=18)
plt.ylabel('Gamma Function',fontsize=18)
plt.legend(loc=1, fontsize=18)

plt.subplot(1,2,2)    
plt.scatter(alpha,g2, c='g', label = 'lambda=1.0')
plt.plot(alpha,g2, c='y')
plt.xlabel('alpha',fontsize=18)
plt.ylabel('Gamma Function',fontsize=18)
plt.legend(loc=1, fontsize=18)
plt.show()
```

6. Plot `log-normal` distribution with parameters $(\mu,\sigma),\ \mu = 0,1,2;\ \sigma= 0.5, 1, 2$

```python
from math import *
import matplotlib.pyplot as plt
import numpy as np
mean=[0,1,2]
sigma=[0.5,1,2]
X=np.linspace(10,15,100)
f=lambda x,a,b: 1/(x*a*sqrt(2*pi))*exp(-((log(x)-b)**2/2*a**2))
for i in range(len(mean)):
    for j in range(len(sigma)):
        for k in range(len(X)):
            if i!=0:
                F=f(k,i,j)
plt.plot(F,X,label="log normal distribution")
plt.xlabel("X values")
plt.ylabel("F(x)")
plt.legend()
plt.show()
```
